## Scaling the dataframe
Scaling the dataset helps to normalise the dataset by transforming them to fall within a similar scale or range. In this case, the Balance and Balance frequency columns has a huge scale difference between them and so scaling the dataset will help to transform each of them into similiar range.

```python
sc = StandardScaler() 
scaled_df = sc.fit_transform(df)
```

```python
scaled_df
```

### Using PCA for Dimensionality reduction

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_df)
pca_df = pd.DataFrame(data=principal_components ,columns=["PCA1","PCA2"])
pca_df
```

### Finding the optimal K value to be used for clutering
## Finding 'k' value by Elbow Method


```python
inertia = []
range_val = range(1, 15)

for i in range_val:
    kmean = KMeans(n_clusters=i)
    kmean.fit_predict(pd.DataFrame(scaled_df))
    inertia.append(kmean.inertia_)

# Using different colors for the plot
plt.plot(range_val, inertia, marker='o', linestyle='-', color='purple')  # You can change the color here
plt.xlabel('Values of K') 
plt.ylabel('Inertia') 
plt.title('The Elbow Method using Inertia') 
plt.grid(True)  # Adding grid for better readability
plt.show()
```

### As the name suggests, the elbow method is a technique to find the optimal K value (Where K stands for number of clusters). Just by seeing the graph, we can say K=4 is the point where it tends to bend or make an elbow joint if compared to an arm and therefore we will be selecting K=4 as our optimal K value.

### Clustering using KMeans

```python
kmeans_model=KMeans(4)
kmeans_model.fit_predict(scaled_df)
pca_df_kmeans= pd.concat([pca_df,pd.DataFrame({'cluster':kmeans_model.labels_})],axis=1)
```

```python
plt.figure(figsize=(6,6))
ax=sns.scatterplot(x="PCA1",y="PCA2",hue="cluster",data=pca_df_kmeans,palette=['red','green','blue','black'])
plt.title("Clustering using K-Means Algorithm")
plt.show()
```

```python
cluster_df = pd.concat([df,pd.DataFrame({'Cluster':kmeans_model.labels_})],axis=1)
cluster_df.head()
```
